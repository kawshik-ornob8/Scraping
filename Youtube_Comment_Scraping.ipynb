```python
def main():
    # Define the path where you want to save the file
    file_path = '/content/drive/MyDrive/Data/Youtube Comments/Comments 1.csv'

    with open(file_path, 'w', newline='', encoding='utf-8') as csv_file:
        # CSV writer setup
        csv_writer = csv.writer(csv_file)
        csv_writer.writerow(['video_id', 'comment'])  # Adding video_id column for context

        # Loop through each video and fetch comments
        for video_id in VIDEO_IDS:
            print(f"Fetching comments for video ID: {video_id}")
            url = f'https://www.googleapis.com/youtube/v3/commentThreads?part=snippet&videoId={video_id}&key={API_KEY}&maxResults={MAX_RESULTS}'
            response = requests.get(url)

            if response.status_code == 200:
                comments_data = response.json()
                process_comments(comments_data, csv_writer, video_id)

                # Handle pagination if there are more comments
                next_page_token = comments_data.get('nextPageToken')
                page_count = 1  # Initialize page counter for each video
                while next_page_token:
                    page_count += 1  # Increment page counter
                    print(f"Fetching page {page_count} of comments for video ID: {video_id}...")
                    url = f'https://www.googleapis.com/youtube/v3/commentThreads?part=snippet&videoId={video_id}&pageToken={next_page_token}&key={API_KEY}&maxResults={MAX_RESULTS}'
                    response = requests.get(url)
                    comments_data = response.json()
                    process_comments(comments_data, csv_writer, video_id)
                    next_page_token = comments_data.get('nextPageToken')
            else:
                print(f"Failed to retrieve comments for video ID {video_id}, status code: {response.status_code}")

def process_comments(data, csv_writer, video_id):
    """Extracts comments and writes them to the CSV file."""
    for item in data['items']:
        comment_info = item['snippet']['topLevelComment']['snippet']
        comment = comment_info['textDisplay']
        csv_writer.writerow([video_id, comment])

if __name__ == "__main__":
    main()
```
